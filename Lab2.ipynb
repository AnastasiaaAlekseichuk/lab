№2.1.9 Напишите программу на Python для отображения значений пределов текущей оси и установки новых значений оси.

```
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.arange(0, 60)
y = 2*x
ax.plot(x,y)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Отрисовка линии')
print('По оси ox: ',ax.get_xlim())
ax.set_xlim(0,100)
print('По оси oy: ',ax.get_ylim())
ax.set_ylim(0,200)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
fig.set_figheight(3)
fig.set_figwidth(7)
langs = ['Java','Python','PHP','JavaScript','C#', 'C++']
students = [22.2,17.6,8.8,8,7.7,6.7]
ax.bar(langs, students)
ax.set_title('Popularity of Programming Language\nWorldwide, Oct 2017 compared to a year ago')
plt.xlabel('Languages')
plt.ylabel('Popularity')
ax.set_xlim(-0.40,5.6)
ax.set_ylim(0,25)
ax.minorticks_on()
ax.grid(which='major',
        color = 'red', 
        linewidth = 1)
ax.grid(which='minor', 
        color = 'grey', 
        linestyle = ':')
ax.set_xticks(np.arange(-0.40,5.6))
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```
