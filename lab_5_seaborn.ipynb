```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv('country_vaccinations 2 (1).csv')
df
```

```
import numpy as np
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(0).astype(np.int64)
df['total_vaccinations'] = df['total_vaccinations'].fillna(0).astype(np.int64)
```

```
df.info()
```

```
df.dropna() 
```

# **Box plots**

```
sns.boxplot(df['total_vaccinations'])
```

```
sns.boxplot(x='total_vaccinations_per_hundred', data=df)
```

**Kernel density estimate**

```
sns.kdeplot(df['total_vaccinations_per_hundred'])
```

```
sns.kdeplot(df['daily_vaccinations'])
```

```
sns.kdeplot(df['daily_vaccinations_raw'])
```

# **Violin plot**

```
sns.violinplot(x='total_vaccinations_per_hundred', y='total_vaccinations', data=df)
```

# **Count plots**

```
sns.countplot(df['total_vaccinations_per_hundred'])
```

# **Heatmap**

```
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
```

# **Facet grid**

```
g=sns.FacetGrid(df, col='country', row='date')
g.map(plt.hist,'total_vaccinations_per_hundred')
```

# **Regplot**

```
import seaborn as sns
sns.regplot(x='total_vaccinations', y='daily_vaccinations', data=df)
```

# **Strip plot**

```
sns.stripplot(x='country', y='daily_vaccinations_raw', data=df)
```

# **Swarm plot**

```
sns.swarmplot(x='country', y='daily_vaccinations_raw', data=df)
```

```
sns.violinplot(x='country', y='daily_vaccinations_raw', data=df, palette='pink')
sns.swarmplot(x='country', y='daily_vaccinations_raw', data=df, color='blue', size=3)
```

# **Pair plot**

```
sns.pairplot(data=df)
```

# **Joint plot**

```
sns.jointplot(x='date', y='total_vaccinations_per_hundred', data=df)
```
